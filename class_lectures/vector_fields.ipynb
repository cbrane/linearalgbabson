```python
import numpy as np
import matplotlib.pyplot as plt

# Define the grid for plotting vector fields within the range (-2, 2)
x, y = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))

# Create a figure for the vector fields
plt.figure(figsize=(15, 5))

# Problem 5, Exercise 1: F = yi - xj
plt.subplot(1, 3, 1)
Fx1 = y
Fy1 = -x
plt.quiver(x, y, Fx1, Fy1)
plt.title(r'Exercise 1: $\mathbf{F} = yi - xj$')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axis('equal')

# Problem 5, Exercise 2: F = (-x, y)
plt.subplot(1, 3, 2)
Fx2 = -x
Fy2 = y
plt.quiver(x, y, Fx2, Fy2)
plt.title(r'Exercise 2: $\mathbf{F} = (-x, y)$')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axis('equal')

# Problem 5, Exercise 3: F = (x^2, x)
plt.subplot(1, 3, 3)
Fx3 = x**2
Fy3 = x
plt.quiver(x, y, Fx3, Fy3)
plt.title(r'Exercise 3: $\mathbf{F} = (x^2, x)$')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axis('equal')

# Show the plots
plt.tight_layout()
plt.show()
```
